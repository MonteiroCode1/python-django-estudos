### Exercicio Sobre Heran√ßa __*basica*__ (bem simples)

````python
from abc import ABC, abstractmethod

class Fucionario(ABC):
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    
    def __str__(self) -> str:
        resultado = f"Nome: {self._nome}\n"
        return resultado
    
    @abstractmethod
    def calcular_salario(self) -> float:
        pass
    
    @property
    def nome(self) -> str:
        return self._nome
    
    @nome.setter
    def nome(self, nome: str) -> None:
        self._nome = nome

#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

class FucionarioCLT(Fucionario):
    
    def calcular_salario(self) -> float:
        return self._salario * 1.1
    
    def __str__(self) -> str:
        resultado = super().__str__() + "\n"
        resultado += f"Salario: {self.calcular_salario():.2f}"
        return resultado

class FucionarioPJ(Fucionario):

    def calcular_salario(self) -> float:
        return self._salario * 1
    
    def __str__(self) -> str:
        resultado = super().__str__() + "\n"
        resultado += f"Salario PJ: {self.calcular_salario()}"
        return resultado
````

````python
pessoasCLT = FucionarioCLT("Guilherme", 1500)
print(pessoasCLT)

pessoasPJ = FucionarioPJ("Mateus", 1500)
print(pessoasPJ)
````

## üß™ Teste esperado (exemplo)

```python

````python
from abc import ABC, abstractmethod

class Conta(ABC):
    def __init__(self, titula: str) -> None:
        self._titula = titula
        self._saldo = 0

    def despositar_valores(self, valor: float) -> None:
        if (valor > 0):
            self._saldo += valor
        else:
            print("Error... O valor deve ser Maior que zero")

    def sacar(self, valor: float) -> None:
        if (valor < 0):
            print("Error... Valor Invalido!")
        elif (valor > self._saldo):
            print("Erro... Saldo insuficiente")
        else:
            self._saldo -= valor
    
    @abstractmethod
    def calcular_taxa(self) -> None:
        pass

    @property
    def saldo(self) -> float:
        return self._saldo
    
#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

class ContaCorrente(Conta):

    def sacar(self, valor: float) -> None:
        if (valor > self._saldo or valor < 0):
            print("Erro... Saldo insuficiente!")
        else:
            self._saldo -= valor
            self.calcular_taxa()


    def calcular_taxa(self) -> None:
        self._saldo -= 5.00

#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

class ContaPoupan√ßa(Conta):

    def calcular_taxa(self) -> None:
        self._saldo *= 1.01


#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
````

````python
class Contato:
    def __init__(self, nome: str, telefone: int) -> None:
        self._nome = nome
        self._telefone = telefone
        self._lista_dos_contatos = []
    
    def pessoa(self) -> list:
        self._lista_dos_contatos.extend(self._nome, self._telefone)
        return self._lista_dos_contatos
    
    def __str__(self) -> str:
        resultado = f"Nome: {self._nome}, Telefone: {self._telefone}"
        return resultado

    @property
    def nome(self) -> str:
        return self._nome
    
    @property
    def telefone(self) -> int:
        return self._telefone

    @nome.setter
    def nome(self, nome: str) -> None:
        self._nome = nome
    
class RepositorioContatos:
    def __init__(self) -> None:
        self._lista_contatos = []
    
    def incluir(self, valor: list) -> None:
        self._lista_contatos.append(valor)
    
    def alterar(self, valor: str) -> None:
        tem = False
        for elemento in self._lista_contatos:
            if (valor in elemento):
                tem = True
                numero = int(input("Digite um Numero Novo: "))
                indice = self._lista_contatos.index(valor)
                self._lista_contatos[indice][1] = numero
        if not tem: ## se der true n√£o tem esse valor| se der False, quer disser que tem
            print(f"o Nome: {valor} n√£o existe nos nossos contatos!")


    def excluir(self, indice: int) -> None:
        if (indice > len(self._lista_contatos) - 1):
            print(f"Coloque um indice de 0 a {len(self._lista_contatos) - 1}")
        else:
            self._lista_contatos.pop(indice)
    
    def __getitem__(self, indice: int) -> list:
        if (indice > len(self._lista_contatos) - 1):
            print(f"para a consulta, digite de 0 a {len(self._lista_contatos) - 1}")
        else:
            return self._lista_contatos[indice]

    @property
    def lista(self) -> list:
        return self._lista_contatos

````

- Exercicio dia 2 - Encapsulamento

````python
class Conta:
    def __init__(self, saldo: float) -> None:
        self._saldo = saldo
    
    def depositar(self, valor: float) -> None:
        if (valor < 0 or valor == 0):
            print("Coloque valores positivos, ou maior que zero!")
        else:
            self._saldo += valor
    
    def sacar(self, valor: float) -> None:
        if (valor > self._saldo or valor < 0):
            print("Saldo insuficiente!")
        else:
            self._saldo -= valor
    
    @property
    def saldo(self) -> float:
        return self._saldo

cidadao = Conta(50.25)
cidadao.depositar(250)
cidadao.sacar(30)
cidadao.saldo
````

````python
class Produto:
    def __init__(self, nome: str, pre√ßo: float) -> None:
        self.nome = nome
        self.pre√ßo = pre√ßo
    
    def aplicar_desconto(self, porcentual: float) -> bool|float:
        valor_desconto = self.pre√ßo * (porcentual / 100)
        pre√ßo_final = self.pre√ßo - valor_desconto
        if (porcentual < 0 or pre√ßo_final < 0):
            return False
        else:
            return pre√ßo_final

mause = Produto("Mause Gamer", 9.99)

mause.aplicar_desconto(200)
````
