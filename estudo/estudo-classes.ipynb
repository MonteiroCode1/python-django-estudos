### Exercicio Sobre HeranÃ§a __*basica*__ (bem simples)

````python
from abc import ABC, abstractmethod

class Fucionario(ABC):
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    
    def __str__(self) -> str:
        resultado = f"Nome: {self._nome}\n"
        return resultado
    
    @abstractmethod
    def calcular_salario(self) -> float:
        pass
    
    @property
    def nome(self) -> str:
        return self._nome
    
    @nome.setter
    def nome(self, nome: str) -> None:
        self._nome = nome

#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

class FucionarioCLT(Fucionario):
    
    def calcular_salario(self) -> float:
        return self._salario * 1.1
    
    def __str__(self) -> str:
        resultado = super().__str__() + "\n"
        resultado += f"Salario: {self.calcular_salario():.2f}"
        return resultado

class FucionarioPJ(Fucionario):

    def calcular_salario(self) -> float:
        return self._salario * 1
    
    def __str__(self) -> str:
        resultado = super().__str__() + "\n"
        resultado += f"Salario PJ: {self.calcular_salario()}"
        return resultado
````

````python
pessoasCLT = FucionarioCLT("Guilherme", 1500)
print(pessoasCLT)

pessoasPJ = FucionarioPJ("Mateus", 1500)
print(pessoasPJ)
````

## ðŸ”¹ Requisitos

### 1ï¸âƒ£ Classe `Conta`
Crie uma classe base chamada `Conta` com:

#### Atributos (protegidos)
- `_titular` â†’ nome do dono da conta
- `_saldo` â†’ saldo da conta (inicia com 0 ou valor informado)

#### MÃ©todos
- `depositar(valor)`
  - SÃ³ permite valores maiores que zero

  - NÃ£o permite saque maior que o saldo
  - Deve aplicar automaticamente a taxa da conta

  - MÃ©todo que serÃ¡ sobrescrito nas classes filhas

  - Property apenas para leitura (nÃ£o pode alterar diretamente)

### 2ï¸âƒ£ Classe `ContaCorrente`
Herda de `Conta` e:

### 3ï¸âƒ£ Classe `ContaPoupanca`
Herda de `Conta` e:
- **NÃ£o cobra taxa**
- Sobrescreve o mÃ©todo `calcular_taxa()`

---

### 4ï¸âƒ£ Polimorfismo
O mÃ©todo `sacar()` deve funcionar de forma diferente dependendo do tipo da conta:
- `ContaCorrente` aplica taxa
- `ContaPoupanca` nÃ£o aplica taxa

---

## ðŸ§ª Teste esperado (exemplo)

```python
cc = ContaCorrente("Gui", 100)
cp = ContaPoupanca("Ana", 100)

cc.sacar(50)
cp.sacar(50)

print(cc.saldo)
print(cp.saldo)


````python
from abc import ABC, abstractmethod

class Conta(ABC):
    def __init__(self, titula: str) -> None:
        self._titula = titula
        self._saldo = 0

    def despositar_valores(self, valor: float) -> None:
        if (valor > 0):
            self._saldo += valor
        else:
            print("Error... O valor deve ser Maior que zero")

    def sacar(self, valor: float) -> None:
        if (valor < 0):
            print("Error... Valor Invalido!")
        elif (valor > self._saldo):
            print("Erro... Saldo insuficiente")
        else:
            self._saldo -= valor
    
    @abstractmethod
    def calcular_taxa(self) -> None:
        pass

    @property
    def saldo(self) -> float:
        return self._saldo
    
#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

class ContaCorrente(Conta):

    def sacar(self, valor: float) -> None:
        if (valor > self._saldo or valor < 0):
            print("Erro... Saldo insuficiente!")
        else:
            self._saldo -= valor
            self.calcular_taxa()


    def calcular_taxa(self) -> None:
        self._saldo -= 5.00

#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

class ContaPoupanÃ§a(Conta):

    def calcular_taxa(self) -> None:
        self._saldo *= 1.01


#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
````
