### Estudo sobre o metodo __str__ de classes

```python
class pessoa:
    def __init__(self, nome: str, idade: int) -> None:
        self._nome = nome
        self._idade = idade
    
    def __str__(self) -> str:
        return f"Nome: {self._nome} idade: {self._idade}"

cidadao = pessoa("Guilherme", 20)
print(cidadao)
```

### Exercicio Sobre HeranÃ§a __*basica*__ (bem simples)

```python
from abc import ABC, abstractmethod

class Fucionario(ABC):
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    
    def __str__(self) -> str:
        resultado = f"Nome: {self._nome}\n"
        return resultado
    
    @abstractmethod
    def calcular_salario(self) -> float:
        pass
    
    @property
    def nome(self) -> str:
        return self._nome
    
    @nome.setter
    def nome(self, nome: str) -> None:
        self._nome = nome

#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

class FucionarioCLT(Fucionario):
    
    def calcular_salario(self) -> float:
        return self._salario * 1.1
    
    def __str__(self) -> str:
        resultado = super().__str__() + "\n"
        resultado += f"Salario: {self.calcular_salario():.2f}"
        return resultado

class FucionarioPJ(Fucionario):

    def calcular_salario(self) -> float:
        return self._salario * 1
    
    def __str__(self) -> str:
        resultado = super().__str__() + "\n"
        resultado += f"Salario PJ: {self.calcular_salario()}"
        return resultado
```

```python
pessoasCLT = FucionarioCLT("Guilherme", 1500)
print(pessoasCLT)

pessoasPJ = FucionarioPJ("Mateus", 1500)
print(pessoasPJ)
```

```python
#exercicio sobre veiculos
class Veiculo:
    def __init__(self, marca: str, velocidade: int) -> None:
        self._marca = marca
        self._velocidade = velocidade

    def acelerar(self) -> int:
        self._velocidade += 10
        return self._velocidade
    
    def __str__(self) -> str:
        resultado = f"Marca: {self._marca}\n"
        resultado += f"Velocidade: {self.acelerar()}Km/h"
        return resultado

class Carro(Veiculo):

    def acelerar(self) -> None:
        self._velocidade += 20
        return self._velocidade
    
    def __str__(self) -> str:
        resultado = super().__str__()
        return resultado

class Moto(Veiculo):

    def acelerar(self) -> None:
        self._velocidade += 30
        return self._velocidade
    
    def __str__(self) -> str:
        resultado = super().__str__()
        return resultado
```

```python
veiculos = [Carro("Estrada", 40), Moto("Honda", 20)]

for v in veiculos:
    v.acelerar()
    v.acelerar()
    print(v)
    print("-" * 20)
```
