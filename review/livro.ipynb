```python
# Herança

class Fucionario:
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    def calcular_bonus(self) -> float:
        aumento = self._salario / 10
        self._salario += aumento
        return self._salario

class Gerente(Fucionario):
    def __init__(self, nome, salario):
        super().__init__(nome, salario)
    
    def calcular_bonus(self):
        aumento = self._salario / 20
        self._salario += aumento
        return self._salario
```

```python
f = Fucionario("Guilherme", 12000.00)
g = Gerente("Klecius", 20000.00)

lista = [f, g]
for e in lista:
    print(e.calcular_bonus())
```

```python
class Fucionario:
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    def calcular_bonus(self) -> float:
        aumento = self._salario / 10
        self._salario += aumento
        return self._salario

class Gerente(Fucionario):
    def __init__(self, nome, salario):
        super().__init__(nome, salario)
    
    def calcular_bonus(self):
        valor = super().calcular_bonus()
        aumento = valor / 10
        self._salario += aumento
        return self._salario 
```

```python
g = Gerente("Gui", 1500)
print(g.calcular_bonus())
```

## Sistema de Pagamentos 🧾

```python
## Exeções
class ValorPagamentoInvalidoError(Exception):
    def __init__(self, mensagem: str) -> None:
        self.mensagem = mensagem
    
    def __str__(self) -> str:
        return self.mensagem
```

```python
## Objetivo: Simule a formas de Pagamento
from abc import ABC, abstractmethod


##class Abstrada
class Pagamento(ABC):
    def __init__(self, valor: float) -> None:
        if valor <= 0:
            raise ValorPagamentoInvalidoError("Erro... Valor Invalido.. deve ser um valor positivo ou maior que Zero!")
        self._valor = valor
    
    @abstractmethod
    def processar_pagamentos(self) -> None:
        pass

##class Filhas
class PagamentoCartao(Pagamento):
    def __init__(self, valor: float) -> None:
        super().__init__(valor)
    
    def processar_pagamentos(self):
        print(f"pagamento de R$ {self._valor} Realizado no cartão!")

##class filhas
class PagamentoPix(Pagamento):
    def __init__(self, valor: float) -> None:
        super().__init__(valor)

    def processar_pagamentos(self):
        print(f"pagamento de R$ {self._valor} realizado via Pix")

##class filhas
class PagamentoBoleto(Pagamento):
    def __init__(self, valor: float) -> None:
        super().__init__(valor)
    
    def processar_pagamentos(self):
        print(f"Boleto gerado no Valor de R$ {self._valor}")
```

```python
# Usando o polimorfismo
pix = PagamentoPix(100.00)
cartao = PagamentoCartao(1.0)
boleto = PagamentoBoleto(30.00)

pagamentos = [pix, cartao, boleto]

for e in pagamentos:
    e.processar_pagamentos()
```
