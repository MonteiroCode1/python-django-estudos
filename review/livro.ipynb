```python
# Herança

class Fucionario:
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    def calcular_bonus(self) -> float:
        aumento = self._salario / 10
        self._salario += aumento
        return self._salario

class Gerente(Fucionario):
    def __init__(self, nome, salario):
        super().__init__(nome, salario)
    
    def calcular_bonus(self):
        aumento = self._salario / 20
        self._salario += aumento
        return self._salario
```

```python
f = Fucionario("Guilherme", 12000.00)
g = Gerente("Klecius", 20000.00)

lista = [f, g]
for e in lista:
    print(e.calcular_bonus())
```

```python
class Fucionario:
    def __init__(self, nome: str, salario: float) -> None:
        self._nome = nome
        self._salario = salario
    def calcular_bonus(self) -> float:
        aumento = self._salario / 10
        self._salario += aumento
        return self._salario

class Gerente(Fucionario):
    def __init__(self, nome, salario):
        super().__init__(nome, salario)
    
    def calcular_bonus(self):
        valor = super().calcular_bonus()
        aumento = valor / 10
        self._salario += aumento
        return self._salario 
```

```python
g = Gerente("Gui", 1500)
print(g.calcular_bonus())
```
