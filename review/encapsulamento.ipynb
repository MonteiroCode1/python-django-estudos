```python
# 📓 Exercícios Práticos de POO
# 1. Encapsulamento: O Cofre do Banco 🔒
# Objetivo: Proteger dados sensíveis e controlar o acesso através de métodos.

# Classe: ContaBancaria

# Atributos: titular (público) e _saldo ou __saldo (privado/protegido).

# Regras: 1. O saldo não pode ser alterado diretamente de fora da classe.
# 2. Criar método depositar(valor).
# 3. Criar método sacar(valor) (verificar se há saldo suficiente).
# 4. Criar um método (ou property) para exibir o saldo atual.

```

```python
class ValorInvalidoError(Exception):
    def __init__(self, mensagem:str) -> None:
        self.mensagem = mensagem
    
    def __str__(self) -> str:
        return self.mensagem

class ContaBancaria:
    def __init__(self, titular:str) -> None:
        self.titular = titular # Publico
        self._saldo = 0 # atributo protegido
    
    def depositar(self, valor:float) -> None:
        if (valor < 0):
            raise ValorInvalidoError("Error... Digite um Valor Positivo...")
        self._saldo += valor
    
    def sacar(self, valor:float) -> None:
        if (valor > self._saldo or valor < 0):
            raise ValorInvalidoError(f"Error... {valor} é maior que Seu saldo atual de {self._saldo:2f}")
        if (valor < 0):
            raise ValorInvalidoError("Error... Digite Valores Positivos!...")
        self._saldo -= valor
    
    @property
    def saldo(self) -> float:
        return self._saldo

```

```python
guilherme = ContaBancaria('Guilherme Monteiro Duarte')
guilherme.depositar(10.00)
```

```python
guilherme.sacar(10)
```

```python
guilherme.saldo
```
