```python
# Herança: Sistema Escolar 🏫
# Objetivo: Reutilizar código de uma classe base para classes mais específicas.

# Classe Base: Pessoa (Atributos: nome, idade).

# Classes Filhas: * Aluno (Atributo extra: matricula).

# Professor (Atributo extra: disciplina).

# Tarefa: Implemente um método apresentar() na classe pai que seja herdado pelas filhas, exibindo os dados de forma organizada.
```

```python
class Pessoa:
    def __init__(self, nome: str, idade: int) -> None:
        self.nome = nome
        self.idade = idade
    
    def apresentar(self):
        print(f'Nome: {self.nome}, idade: {self.idade}')

class Aluno(Pessoa):
    def __init__(self, nome: str, idade: int, matricula: int) -> None:
        super().__init__(nome, idade)
        self.matricula = matricula
    def apresentar(self):
        print(f"Nome: {self.nome}, idade: {self.idade}, matricula: {self.matricula}")

class Professor(Pessoa):
    def __init__(self, nome: str, idade: int, disciplina: str) -> None:
        super().__init__(nome, idade)
        self.disciplina = disciplina
    
    def apresentar(self):
        print(f"Nome: {self.nome}, idade: {self.idade}, disciplina: {self.disciplina}")

gente = Pessoa("Guilherme", 20)
estudante = Aluno("Igor", 36, 2345)
lider = Professor("Mauricio", 39, 'cabeleleiro')

lista = [gente, estudante, lider]

for e in lista:
    e.apresentar()
```
