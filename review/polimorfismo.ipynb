```python
# 3. Polimorfismo: Meios de Transporte 🚗✈️
# Objetivo: Fazer com que objetos diferentes respondam ao mesmo comando de formas distintas.

# Classe Pai: Veiculo com o método mover().

# Classes Filhas: Carro, Aviao, Barco.

# Tarefa: Sobrescreva o método mover() em cada uma:

# Carro: "Acelerando na rodovia."

# Avião: "Decolando para as nuvens."

# Barco: "Navegando pelo oceano."

# Desafio: Crie uma lista com instâncias dos três e use um for para chamar o mover() de todos.
```

```python
class Veiculo:
    def mover(self):
        print('Movendo nas Estradas')

class Carro(Veiculo):
    def mover(self):
        print("Acelerando na rodovia.")

class Aviao(Veiculo):
    def mover(self):
        print("Decolando para as nuvens.")

class Barco(Veiculo):
    def mover(self):
        print("Navegando pelo oceano.")

transporte = Veiculo()
Saveira = Carro()
ajato = Aviao()
titanic = Barco()

instacias = [transporte, Saveira, ajato, titanic]

for e in instacias:
    e.mover()
```
